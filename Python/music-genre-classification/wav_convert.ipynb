Import libraries

```python
import numpy as np, sys, os, glob
from pydub import AudioSegment
```

Convert .mp3 files to .wav (only a 30-second slice in the middle of the song)

```python
def wav_convert(audio_dir):
    os.chdir(audio_dir)

    for mp3_file in glob.glob('*.mp3'):
    
        print('Converting ' + mp3_file)
    
        # only get the middle 30 seconds (slicing in miliseconds)
        cut = AudioSegment.from_mp3(mp3_file)[50*1000:80*1000]
        # rename as .wav file
        wav_filename = os.path.splitext(os.path.basename(mp3_file))[0] + '.wav'
    
        print('Storing .wav file: ' + wav_filename)
        # convert cut to .wav
        cut.export(wav_filename, format='wav')
```

```python
######## train data #########
audio_dir = './Data/train/'
wav_convert(audio_dir)
```

```python
######## test data #########
audio_dir = './Data/test/'
wav_convert(audio_dir)
```
